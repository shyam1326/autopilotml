# Stage 2: Data Cleaning

```python
df = preprocessing(dataframe=df, label_column = 'Survived', missing={'type': 'impute',
                                                                    'drop_columns': False, 
                                                                    'threshold': 0.25, 
                                                                    'strategy_numerical': 'knn',
                                                                    'strategy_categorical': 'most_frequent',
                                                                    'fill_value': None},
                                                                    outlier={'method': 'None',
                                                                    'zscore_threshold': 3,
                                                                    'iqr_threshold': 1.5,
                                                                    'Lc': 0.05, 
                                                                    'Uc': 0.95,
                                                                    'cap': False})
df
```

# Stage 3: Transformation

```python
df
```

```python
df, encoder = transformation(dataframe=df, label_column='Survived', type = 'ordinal',target_transform = False, cardinality = True, Cardinality_threshold = 30)
df.head()
```

```python
training(dataframe=df, label_column='Survived', model_name='RandomForestClassifier', problem_type='Classification', 
        target_scaler=None, test_split =0.15, hypertune=True, n_epochs=5)
```

# Regression Usecase

```python
# Load data
df = load_data(path = "/Volumes/Projects/autopilotml/autopilotml/dataset/Power_Output.csv" , csv=True)
df.head()
```

```python
df.shape
```

```python
df = preprocessing(dataframe=df, label_column = 'Electrical Energy Output', missing={'type': 'impute',
                                                                                    'drop_columns': False, 
                                                                                    'threshold': 0.25, 
                                                                                    'strategy_numerical': 'knn',
                                                                                    'strategy_categorical': 'most_frequent',
                                                                                    'fill_value': None},
                                                                                    outlier={'method': 'iqr',
                                                                                    'zscore_threshold': 3,
                                                                                    'iqr_threshold': 1.5,
                                                                                    'Lc': 0.25, 
                                                                                    'Uc': 0.75,
                                                                                    'cap': False})
df
```

```python
df, encoder = transformation(dataframe=df, label_column='Electrical Energy Output', type = 'ordinal',target_transform = False, cardinality = True, Cardinality_threshold = 30)
df.head()
```

```python
df, scaler, target_scaler = scaling(df, label_column= 'Electrical Energy Output', type = 'standard', target_scaling = True)
df.head()
```

```python
df, selector = feature_selection(dataframe=df, label_column='Electrical Energy Output', estimator='RandomForestRegressor', type='rfe', max_features=3, min_features=2, scoring= 'accuracy', 
                    cv=5)
df
```

```python
model = training(dataframe=df, label_column='Electrical Energy Output', model_name='RandomForestRegressor', problem_type='Regression', 
        target_scaler=target_scaler, test_split =0.15, hypertune=True, n_epochs=5)
```

```python
df.iloc[0:1,0:3]
```

```python
pred = model.predict(df.iloc[0:1,0:3])
pred
```

```python
target_scaler.inverse_transform(pred.reshape(-1,1)) # The ground truth is 463.26
```
