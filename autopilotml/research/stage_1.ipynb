# Import Libraries

```python
import os
os.chdir('/Volumes/Projects/autopilotml/autopilotml/')

import warnings
warnings.filterwarnings("ignore")

from autopilotml import load_data, preprocessing, transformation, scaling, feature_selection, training
```

# Load data from local file

```python

# Load data
df = load_data(path = "/Volumes/Projects/autopilotml/autopilotml/dataset/titanic_train.csv" , csv=True, header=0)
df.head()
```

# Load data from Database

```python
# df = load_database(database_type='sqlite', sqlite_db_path = 'database.db', query='select * from employee_table')
# df.head()
```

# Stage 2: Data Cleaning

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df = preprocessing(dataframe=df, label_column = 'Survived', missing={'type': 'impute',
                                                                    'drop_columns': False, 
                                                                    'threshold': 0.25, 
                                                                    'strategy_numerical': 'knn',
                                                                    'strategy_categorical': 'most_frequent',
                                                                    'fill_value': None},
                                                                    outlier={'method': 'None',
                                                                    'zscore_threshold': 3,
                                                                    'iqr_threshold': 1.5,
                                                                    'Lc': 0.05, 
                                                                    'Uc': 0.95,
                                                                    'cap': False})
df
```

```python
df.isnull().sum()
```

# Stage 3: Transformation

```python
df
```

```python
df, scaler = transformation(dataframe=df, label_column='Survived', type = 'ordinal',target_transform = False, cardinality = True, Cardinality_threshold = 0.3)
df.head()
```

# Scaling

```python
df, scaler = scaling(df, label_column= 'Survived', type = 'standard', target_scaling = False)
```

```python
df.head()
```

# Feature Selection

```python
from sklearn.svm import SVC, SVR
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
```

```python
df, selector = feature_selection(dataframe=df, label_column='Survived', estimator=RandomForestClassifier(), type='rfe', max_features=10, min_features=2, scoring= 'accuracy', 
                    cv=5)
```

```python
df.head()
```

# Training

```python
training(dataframe=df, label_column='Survived', model_name='SVC', problem_type='Classification', 
        target_scaler=None, test_split =0.15, hypertune=True, n_epochs=5)
```
